```python
import numpy as np
def policy(q_values, epsilon):
    random_number = np.random.rand()
    if random_number < epsilon:
        return np.random.randint(0,5)
    else:
        return np.argmax(q_values)
```

```python
def reward(action):
    if action == 0:
        return Gaussian(2,1)
    elif action == 1:
        return Coin_toss([5,-6])
    elif action == 2:
        return Poisson(2)
    elif action == 3:
        return Exponential(3)
    elif action == 4:
        return Crazy_Button()
```

```python
def Gaussian(mean, std_dev):
    return np.random.normal(mean, std_dev)
```

```python
def Coin_toss(outcomes):
    return np.random.choice(outcomes)
```

```python
def Poisson(lambda_poisson):
    return np.random.poisson(lambda_poisson)
```

```python
def Exponential(mean):
    return np.random.exponential(mean)
```

```python
def Crazy_Button():
    choices = [Gaussian(2,1), Coin_toss([5,-6]), Poisson(2), Exponential(3)]
    return np.random.choice(choices)
```

```python
q_values = [0,0,0,0,0]
occurences = [0,0,0,0,0]
epsilon = 0.01
reward_hist = []
for i in range(0, 1000):
    episodic_reward = 0
    for j in range(0, 100):
        action = policy(q_values, epsilon)
        occurences[action] += 1
        current_reward = reward(action)
        episodic_reward += current_reward
        q_values[action] += (current_reward - q_values[action])/occurences[action]
    reward_hist.append(episodic_reward)
    
print(q_values)
print(occurences)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(reward_hist, label = "0.01")
plt.legend()
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("Rewards")
plt.show()
```
